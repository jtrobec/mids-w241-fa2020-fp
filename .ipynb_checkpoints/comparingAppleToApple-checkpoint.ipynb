```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 

#to import pictures as numpy
import PIL
from PIL import Image
from matplotlib import pyplot
```

```python
#op is where original picture is located at
OP = "data/B/exper1/pic/t2/"

#cp is where cropped picture will be saved
CP = OP + "cp/"

#the reference image
rp = OP + "a4t2.JPG"

a0 = Image.open(rp)
print(a0.format) #it is jpeg
print(a0.size) #size of image
print(a0.mode) #it is RGB
```

```python
image = cv2.imread(rp)
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
edged = cv2.Canny(image, 10, 250) 
(cnts, _) = cv2.findContours(edged.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE) 


idx = 0
for c in cnts: 
    x,y,w,h = cv2.boundingRect(c) 
    if w>50 and h>50: 
        idx+=1 
        new_img=image[y:y+h,x:x+w] 
        cv2.imwrite(str(idx) + '.png', new_img) 
```
