### For Nobu's photos

```python
import os
import cv2
import numpy as np

folder = 'data/N/'

for i in range(1, 11, 1):
    day = 't'+ str(i)
    path = folder + 'experiment3/' + day + '/banana'
    idx = 0
    for filename in os.listdir(path):
        if filename.endswith('.jpg'):
            num = ''
            j = 0
            while True:
                if filename[j] == 't':
                    break
                else:
                    num = num + filename[j]
                    j += 1
            print(filename)
            print(num)
            #############################################
            # After selecting the correct HSV
            #############################################
            img = cv2.imread(path + '/' + filename)
            #print(img)
            #convert to HSA
            imgHSV = cv2.cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
            #select the filtered image
            lower = np.array([21,36,44])
            upper = np.array([55,255,255])
            #filtered image
            mask = cv2.inRange(imgHSV, lower, upper)
            #using filtered images to create new image 
            #this is adding two images together 
            imgResult = cv2.bitwise_and(img,img, mask = mask)
 
            # =============================================================================
            # img = imgResult.copy() 
            # imgContour = imgResult.copy()
            # =============================================================================

            imgGray = cv2.cvtColor(imgResult, cv2.COLOR_BGR2GRAY)
            #sigma, the higher the value, the more blur you get
            imgBlur = cv2.GaussianBlur(imgGray,(7,7),1)

            (cnts, _) = cv2.findContours(imgBlur, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE) 
        #    idx = 0 
            for c in cnts: 
                area = cv2.contourArea(c)
                if area > 1000:
                    x,y,w,h = cv2.boundingRect(c) 
                    if w>400 and h>400:
                        print('check4')
                        idx+=1 
                        new_img=imgResult[y:y+h,x:x+w] 
                        print('banana_images/N/' +day + '/'+ num + '.png')
                        cv2.imwrite('banana_images/N/' +day + '/'+ num + '.png', new_img) 
```
